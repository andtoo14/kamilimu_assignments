Create an in-memory SQLite database.
Load the necessary datasets.
Write the datasets to the in-memory database.

```python
import sqlite3
import pandas as pd

# Load datasets
athletes = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/athlete_events.csv")
regions = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/noc_regions.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep='\t')


# Connect to SQLite in-memory DB
conn = sqlite3.connect(":memory:")

# Write DataFrames to SQL tables
athletes.to_sql("athletes_table", conn, index=False, if_exists="replace")
regions.to_sql("regions_table", conn, index=False, if_exists="replace")
sales.to_sql("sales_table", conn, index=False, if_exists="replace")
```

**SECTION ONE:**

Count the total number of medals won by each country and show the top 5:

```python
print("1. Top 5 Countries by Medals:")
query1 = """
SELECT
    r.region,
    COUNT(a.Medal) AS total_medals
FROM athletes_table a
JOIN regions_table r ON a.NOC = r.NOC
WHERE a.Medal IS NOT NULL
GROUP BY r.region
ORDER BY total_medals DESC
LIMIT 5;
"""
result1 = pd.read_sql_query(query1, conn)
display(result1)
```

Calculate the average age of athletes who won a Gold medal:

```python
print("\n2. Average Age of Gold Medalists:")
query2 = """
SELECT
    AVG(Age) AS average_age_gold_medalists
FROM athletes_table
WHERE Medal = 'Gold';
"""
result2 = pd.read_sql_query(query2, conn)
display(result2)
```

How many distinct events are there in each sport?

```python
print("\n3. Distinct Events per Sport:")
query3 = """
SELECT
    Sport,
    COUNT(DISTINCT Event) AS distinct_events
FROM athletes_table
GROUP BY Sport
ORDER BY distinct_events DESC;
"""
result3 = pd.read_sql_query(query3, conn)
display(result3)
```

Show all athletes from the United States (NOC = 'USA'):

```python
print("\n4. Athletes from the United States (USA):")
query4 = """
SELECT
    *
FROM athletes_table
WHERE NOC = 'USA'
LIMIT 10; -- Limiting to 10 for brevity
"""
result4 = pd.read_sql_query(query4, conn)
display(result4)
```

Count how many medals were awarded each year:

```python
print("\n5. Medals Awarded Each Year:")
query5 = """
SELECT
    Year,
    COUNT(Medal) AS total_medals
FROM athletes_table
WHERE Medal IS NOT NULL
GROUP BY Year
ORDER BY Year;
"""
result5 = pd.read_sql_query(query5, conn)
display(result5)
```

Find all athlete records where height or weight is missing:

```python
print("\n6. Athlete Records with Missing Height or Weight:")
query6 = """
SELECT
    *
FROM athletes_table
WHERE Height IS NULL OR Weight IS NULL
LIMIT 10; -- Limiting to 10 for brevity
"""
result6 = pd.read_sql_query(query6, conn)
display(result6)
```

Replace the missing height with the average athlete height:

```python
print("\n7. Demonstrating how to get the average height to replace missing values:")
query7_avg_height = """
SELECT AVG(Height) FROM athletes_table WHERE Height IS NOT NULL;
"""
avg_height = conn.execute(query7_avg_height).fetchone()[0]
print(f"Average Athlete Height (for replacement): {avg_height}")

# To actually update the table in SQL:
print("\nTo actually replace missing height in the SQL table, you would run an UPDATE query.")
```

Return total sales value per item:

```python
print("\n8. Total Sales Value per Item:")
query8 = """
SELECT
    item_name,
    SUM(CAST(REPLACE(item_price, '$', '') AS REAL)) AS total_sales_value
FROM sales_table
GROUP BY item_name
ORDER BY total_sales_value DESC;
"""
result8 = pd.read_sql_query(query8, conn)
display(result8)
```

Show the top 5 records with the highest item_price:

```python
print("\n9. Top 5 Records with Highest Item Price:")
query9 = """
SELECT
    *
FROM sales_table
ORDER BY CAST(REPLACE(item_price, '$', '') AS REAL) DESC
LIMIT 5;
"""
result9 = pd.read_sql_query(query9, conn)
display(result9)
```

How many unique customer orders are there? (Assume each order_id is a customer.)

```python
print("\n10. Number of Unique Customer Orders:")
query10 = """
SELECT
    COUNT(DISTINCT order_id) AS unique_customer_orders
FROM sales_table;
"""
result10 = pd.read_sql_query(query10, conn)
display(result10)
```

**SECTION TWO:**

Find countries with high-performing athletes in the Olympics. Use at least JOIN, NESTED QUERY, CASE, and optionally WITH.

For each country:

-Count the number of athletes who won at least one medal.

-Determine the average age of those medalists.

-Create a new column called performance:

'High' if average age is below 25

'Medium' if between 25 and 30

'Low' if above 30

```python
query = """
WITH Medalists AS (
    SELECT
        a.ID,
        a.Name,
        a.Team,
        a.Age,
        r.region AS Country,
        a.Medal
    FROM
        athletes_table a
    JOIN
        regions_table r ON a.NOC = r.NOC
    WHERE
        a.Medal IS NOT NULL
    GROUP BY a.ID, a.Name, a.Team, a.Age, r.region, a.Medal
)
SELECT
    Country,
    COUNT(DISTINCT ID) AS NumberOfMedalists,
    AVG(Age) AS AverageAgeOfMedalists,
    CASE
        WHEN AVG(Age) < 25 THEN 'High'
        WHEN AVG(Age) BETWEEN 25 AND 30 THEN 'Medium'
        ELSE 'Low'
    END AS Performance
FROM
    Medalists
GROUP BY
    Country
ORDER BY
    NumberOfMedalists DESC;
"""

result_df = pd.read_sql_query(query, conn)
display(result_df)
```
